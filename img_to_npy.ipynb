```python
import numpy as np
import os
import cv2
```

```python

path_train_glioma = os.path.join("agyikepek_4_osztaly", "Training", "glioma")
path_train_meningioma = os.path.join("agyikepek_4_osztaly", "Training", "meningioma")
path_train_notumor = os.path.join("agyikepek_4_osztaly", "Training", "notumor")
path_train_pituitary = os.path.join("agyikepek_4_osztaly", "Training", "pituitary")

path_test_glioma = os.path.join("agyikepek_4_osztaly", "Testing", "glioma")
path_test_meningioma = os.path.join("agyikepek_4_osztaly", "Testing", "meningioma")
path_test_notumor = os.path.join("agyikepek_4_osztaly", "Testing", "notumor")
path_test_pituitary = os.path.join("agyikepek_4_osztaly", "Testing", "pituitary")

label_map = {
    "glioma": 0,
    "meningioma": 1,
    "notumor": 2,
    "pituitary": 3
}

def img_to_npy(PATH, label, img_size, data_type):
    images = []
    labels = []
    
    for file in os.listdir(PATH):
        fname = os.path.join(PATH, file)
        img = cv2.resize(cv2.imread(fname, cv2.IMREAD_COLOR), img_size)  # Read as RGB and resize
        images.append(img)
        labels.append(label_map[label])
    
    images = np.array(images, dtype=np.uint8)  # Ensure dtype is uint8 for raw pixel values
    labels = np.array(labels, dtype=np.int32)
    
    np.save(f"X_{img_size[0]}_{data_type}.npy", images)
    np.save(f"y_{data_type}.npy", labels)
    
    print(f"X_{img_size[0]}_{data_type}.npy saved")
    print(f"y_{data_type}.npy saved")

image_sizes = [(32, 32), (64, 64), (128, 128), (256, 256)]

for img_size in image_sizes:
    img_to_npy(path_train_glioma, "glioma", img_size, "train")
    img_to_npy(path_train_meningioma, "meningioma", img_size, "train")
    img_to_npy(path_train_notumor, "notumor", img_size, "train")
    img_to_npy(path_train_pituitary, "pituitary", img_size, "train")
    
    img_to_npy(path_test_glioma, "glioma", img_size, "test")
    img_to_npy(path_test_meningioma, "meningioma", img_size, "test")
    img_to_npy(path_test_notumor, "notumor", img_size, "test")
    img_to_npy(path_test_pituitary, "pituitary", img_size, "test")
#1m54s
```
